```python
from ultralytics import YOLO 
```

```python
# your yolo model paths goes here
# for the first time it will download the model
yolo_model_path = "yolo model path"
model = YOLO("C:/Users/Darkness/Developments/quantbit/yolo11n.pt")
```

```python
data_dir_path = "data dir path..." # the dataset downloaded from the above provided link
epochs = 10
batch = 16


model.train(
    data="C:/Users/Darkness/Developments/quantbit/fire and smoke.v3i.yolov11/data.yaml",  
    epochs=epochs,                 
    imgsz=640,                
    batch=batch,                
    project="fire_smoke_detection",  
    name="yolov8n_fire_smoke"       
)
```

## Confusion matrix

```python
metrics = model.val()
```

```python
confusion_matrix = metrics.confusion_matrix.matrix

confusion_matrix
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

disp = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix.astype(int))
disp.plot(cmap='Blues')
plt.show()
```
